### 6. To unrar the RAR files:

```python
for rar_file in rar_files:
    print(f"Extracting: {rar_file}")
    rar_file_path = os.path.join(current_directory, rar_file)
    print(f"Rarfile path: {rar_file_path}")
    with rarfile.RarFile(rar_file_path) as rf:
        rf.extractall("temp")
    print("Extraction has completed successfully")
```

### 7. To get the folders list in the extracted directory:

```python
list_folders = [f for f in os.listdir("temp")]
print(list_folders)
```

### 8. To check the YAML File in the each folder on the folders list:

Then I converting all the .yaml/.yml file to .csv file.

```python
for dir in list_folders:
    print("Started to checking the YAML file in the current folder")
    dir_path =  os.path.join("temp", dir)
    for file in os.listdir(dir_path):
        if file.endswith(".yaml") or  file.endswith(".yml"):
            yaml_path = os.path.join(dir_path, file)

            with open(yaml_path, "r") as f:
                yaml_data = yaml.safe_load(f)
            
            df = pd.DataFrame(yaml_data)
            csv_file = os.path.join("csv_output", file.replace(".yaml", ".csv").replace(".yml", ".csv"))
            df.to_csv(csv_file, index=False)
            print(f"CSV file saved in: {csv_file}")
            csv_files.append(csv_file)
```

```python
if csv_files:
    print(csv_files)
```

```python
if csv_files:
    first = True
    print(csv_files)
    with open("dataset.csv", "w", encoding="utf-8", newline="") as fout:
        for cf in csv_files:
            if first:
                with open(cf, encoding="utf-8") as fin:
                    fout.write(fin.read())
                first = False
            else:
                with open(cf, encoding="utf-8") as fin:
                    next(fin)
                    fout.write(fin.read())
    print(f"The file combined to ðŸ¤ž {fout} ðŸ˜")
else:
    print("No CSV file found to combine! ðŸ˜¢")
```

## 2. Cumulative Return Over Time:
Objective:

Show the cumulative return of each stock from the beginning of the year to the end.
Reason: The cumulative return is an important metric to visualize overall performance and growth over time. This helps users compare how different stocks performed relative to each other.

Metrics:

For each stock, calculate the cumulative return by applying a running total of daily returns.
Plot a line chart for the top 5 performing stocks (based on cumulative return) over the course of the year.

Visualization:

Cumulative Return for Top 5 Performing Stocks: A line chart displaying cumulative returns for each stock over the year (increasing trend indicates positive performance).

#### To sort the date by ticker and date

```python
dataset2 = dataset2.sort_values(by=['ticker', 'date'])
dataset2
```

#### To calculate the daily return

The formula is,

    Daily Return = close - previous close/previous close

```python
dataset2['daily_return'] = dataset2.groupby('ticker')['close'].pct_change()
dataset2
```

#### To calculate the cumulative return

The formula is,

    Cumulative return = running multiplication of daily return
    (1+r1)(1+r2)(1+r3)...-1

```python
dataset2['cumulative_return']=(1+dataset2['daily_return']).groupby(dataset2['ticker']).cumprod()-1
dataset2
```

#### To find the top 5 performing stock

```python
final_cumulative = dataset2.groupby('ticker')['cumulative_return'].last()
final_cumulative
```

```python
top_5 = final_cumulative.sort_values(ascending=False).head(5)
print("Top performming Stocks\n", top_5)
```

#### To plot the top 5 cumulative return:

```python
plt.figure(figsize=(12,6))
plt.title("Cumulative Return of Top 5 Performing Stocks!")
for ticker in top_5_tickers:
    stock_data = top_5_dataset[top_5_dataset['ticker'] == ticker]
    plt.plot(stock_data['date'], stock_data['cumulative_return'], label=ticker)
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Top 5 Gainers and Losers (Month-wise):

Objective: 

Provide monthly breakdowns of the top-performing and worst-performing stocks.

Reason: 

This analysis will allow users to observe more granular trends and understand which stocks are gaining or losing momentum on a monthly basis.

Metrics:

Group the stock data by month and calculate the monthly return for each stock.
For each month, identify the top 5 gainers and top 5 losers based on percentage change.
Create a dashboard-style visualization with 5 charts showing top gainers and losers for each month (12 months total).

Visualization:

Top 5 Gainers and Losers by Month: Create a set of 12 bar charts for each month showing the top 5 gainers and losers based on percentage return.

#### To load the dataset

```python
dataset5 = pd.read_sql("SELECT * FROM main", con)
dataset5
```

#### To intialize some variables for future use

```python
TARGET_YEAR = None
TOP_N = 5
MIN_TICKERS_PER_MONTH = 8
ANNOTATE_BAR = True
```

#### Basic validation with dataset

```python
required_cols = {'ticker', 'date', 'close'}
missing = required_cols - set(dataset5.columns)
missing
```

#### To change the datatype of the date column

```python
dataset5['date'] = pd.to_datetime(dataset5['date'], errors="coerce")
dataset5.info()
```

```python
dataset5 = dataset5.sort_values(['ticker', 'date']).reset_index(drop=True)
dataset5
```

#### To find the Target year

```python
if TARGET_YEAR is None:
    TARGET_YEAR = int(dataset5['date'].dt.year.max())
    TARGET_YEAR

dataset5_year = dataset5[dataset5['date'].dt.year == TARGET_YEAR].copy()
dataset5_year
```

#### To easy ordering, need to build monthly key and also a string for filenames

```python
dataset5_year['month_period'] = dataset5_year['date'].dt.to_period('M')
dataset5_year['month_str'] = dataset5_year["month_period"].astype(str)
dataset5_year
```

#### To calculate the monthly return per ticker

```python
monthly = (
    dataset5_year.groupby(['ticker', 'month_period'], sort=False)
    .agg(
        first_close =("close", "first"),
        last_close = ("close", "last"),
        first_date =("date", "first"),
        last_date =("date", "last")
    )
    .reset_index()
)

monthly['monthly_return'] = (monthly["last_close"] / monthly["first_close"] - 1.0) * 100.0
monthly['month_str'] = monthly["month_period"].astype(str)
monthly
```

#### RANKING TOP GAINERS & LOSERS BY MONTH

#### To get the unique months

```python
all_rank_rows = []

unique_months = monthly["month_period"].unique()
unique_months
```

#### To prepare the date in month wise

```python
for m in unique_months:
    m_dataset5 = monthly[monthly['month_period'] == m].copy()

    # To ensure the enough tickers to rank
    if len(m_dataset5) < MIN_TICKERS_PER_SECTOR:
        print(f"Skipping {str(m)}: only {len(m_dataset5)} tickers (min required {MIN_TICKERS_PER_SECTOR}).")
        continue

    # Top N Gainers
    top_gainers = m_dataset5.nlargest(TOP_N, "monthly_return")
    
    # Top N losers
    top_losers = m_dataset5.nsmallest(TOP_N, "monthly_return")

    # To save the per-month tables
    month_tag = str(m)
    month_dir = OUTPUT_DIR / month_tag
    month_dir.mkdir(exist_ok=True)

    gainers_csv = month_dir / f"{month_tag}_top{TOP_N}_gainers.csv"
    losers_csv = month_dir / f"{month_tag}_top{TOP_N}_losers.csv"

    top_gainers.sort_values("monthly_return", ascending=True).to_csv(gainers_csv, index=False)
    top_losers.sort_values("monthly_return", ascending=True).to_csv(losers_csv, index=False)

    # To collect tidy rows for a single combined CSV later

    tg = top_gainers.assign(month=month_tag, kind="gainer").sort_values("monthly_return", ascending=False)
    tl = top_losers.assign(month=month_tag, kind="loser").sort_values("monthly_return", ascending=True)
    all_rank_rows.append(tg[['month', 'ticker', 'monthly_return']].assign(kind="gainer"))
    all_rank_rows.append(tl[["month", "ticker", "monthly_return"]].assign(knid="loser"))

    # To plot for the gainers

    fig_g, ax_g = plt.subplots(figsize=(10, 6))
    g_plot = top_gainers.sort_values("monthly_return", ascending=False)
    ax_g.bar(g_plot['ticker'], g_plot['monthly_return'])
    ax_g.set_title(f"Top {TOP_N} Gainers - {month_tag} ({TARGET_YEAR})")
    ax_g.set_xlabel("Ticker")
    ax_g.set_ylabel("Monthly Return (%)")
    ax_g.tick_params(axis='x', rotation=45)
    ax_g.grid(axis='y', linestyle='--', alpha=0.5)
    fig_g.tight_layout()
    fig_g.savefig(OUTPUT_DIR / f"{month_tag}_top{TOP_N}_gainers.png", dpi=300)
    plt.close(fig_g)

    # To plot for the losers

    fig_l, ax_l = plt.subplots(figsize=(10, 6))
    l_plot = top_losers.sort_values("monthly_return", ascending=True)
    ax_l.bar(g_plot['ticker'], g_plot['monthly_return'])
    ax_l.set_title(f"Top {TOP_N} Losers - {month_tag} ({TARGET_YEAR})")
    ax_l.set_xlabel("Ticker")
    ax_l.set_ylabel("Monthly Return (%)")
    ax_l.tick_params(axis='x', rotation=45)
    ax_l.grid(axis='y', linestyle='--', alpha=0.5)
    fig_l.tight_layout()
    fig_l.savefig(OUTPUT_DIR / f"{month_tag}_top{TOP_N}_losers.png", dpi=300)
    plt.close(fig_l)

if all_rank_rows:
    combined = pd.concat(all_rank_rows, ignore_index=True)
    combined = combined.rename(columns={"monthly_return":"monthly_return_%"})

    # Order by RANK

    combined['month'] = pd.PeriodIndex(combined['month'], freq='M').astype(str)
    combined = combined.sort_values(['month', 'kind', 'monthly_return_%'], ascending=[True, True, False])
    combined.to_csv(OUTPUT_DIR / f"monthly_top{TOP_N}_gainers_losers{TARGET_YEAR}.csv", index=False)

combined

#m_dataset5
```
